````python
# Install llama-cpp-python if not already installed
# !pip install llama-cpp-python

from llama_cpp import Llama


llm = Llama(
    model_path="gguf_model/unsloth.q4_k_m.gguf",
    n_ctx=2048,
    n_gpu_layers=-1,
    verbose=True # Explicitly enable verbose logging
)

print("✅ GGUF model loaded successfully!")
````

````python
# Test code to generate unit tests for
user_code_input = """
def find_max(numbers):
    max_val = 0
    for num in numbers:
        if num > max_val:
            max_val = num
    return max_val
"""

user_problem_description = "A function designed to find the maximum integer in a list of numbers."

# 1. Define the exact same template used during training!
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
"""

# 2. Format the variables identically
complex_input = f"Problem Description:\n{user_problem_description}\n\nCode to Test:\n{user_code_input}"
instruction = "Write a comprehensive Python unit test suite for the provided code."

# 3. Create the final prompt
prompt = alpaca_prompt.format(instruction, complex_input)

# Use standard chat formatting
messages = [
    {"role": "system", "content": "You are a helpful coding assistant that writes Python unit tests."},
    {"role": "user", "content": f"Write a comprehensive Python unit test suite for this code.\n\nProblem Description:\n{user_problem_description}\n\nCode to Test:\n{user_code_input}"}
]

# Use create_chat_completion which applies the correct `<|im_start|>` ChatML tags
output = llm.create_chat_completion(
    messages=messages,  # Changed from prompt=prompt
    max_tokens=512,
    temperature=0.7,
    top_p=0.9
)


print("Generated Unit Tests:")
print("=" * 50)
print(output["choices"][0]["message"]["content"].strip())
````
