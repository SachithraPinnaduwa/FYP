# Inference with GGUF Model

This notebook focuses entirely on running inference using the `llama-cpp-python` package. 

Before running this notebook, ensure you have successfully compiled and converted your model by running the following command in the main `FYP` folder:
```bash
python llama.cpp/convert_hf_to_gguf.py ./merged_model_16bit --outfile gguf_model/unsloth.Q8_0.gguf --outtype q8_0
```

````python
# Install llama-cpp-python if not already installed
# !pip install llama-cpp-python

from llama_cpp import Llama

# Load the manually converted GGUF model
llm = Llama(
    model_path="gguf_model/unsloth.Q8_0.gguf",
    n_ctx=2048,
    n_gpu_layers=-1  # Use all GPU layers (-1 = all, 0 = CPU only)
)

print("✅ GGUF model loaded successfully!")
````

````python
# Test code to generate unit tests for
user_code_input = """
def find_max(numbers):
    max_val = 0
    for num in numbers:
        if num > max_val:
            max_val = num
    return max_val
"""

user_problem_description = "A function designed to find the maximum integer in a list of numbers."

# Format the prompt
prompt = f"""Write a comprehensive Python unit test suite for the provided code.

Problem Description:
{user_problem_description}

Code to Test:
{user_code_input}"""

# Generate unit tests using the GGUF model
output = llm(
    prompt,
    max_tokens=512,
    temperature=0.7,
    top_p=0.9,
    echo=False
)

print("Generated Unit Tests:")
print("=" * 50)
print(output["choices"][0]["text"])
````
